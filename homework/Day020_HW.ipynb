```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
'''
Your Code
'''
trlist = table.find_all('tr', attrs = {'style': 'color:Black;'})
name = ''
dict = {}
for idx, td in enumerate(trlist):
    if idx % 12 == 0:
        name = td.find('td', attrs = {'style': 'white-space:nowrap;'}).string
        date = td.find_all('td')[2].string
        value = td.find_all('td')[3].string
        dict[name] = {date: value}
    else:
        date = td.find_all('td')[0].string
        value = td.find_all('td')[1].string
        dict[name][date] = value
print(dict)
```
