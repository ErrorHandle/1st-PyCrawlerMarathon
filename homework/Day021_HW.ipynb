## 1. 取出今天所有的新聞

```python
import requests
from bs4 import BeautifulSoup
import time
from selenium import webdriver

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "lxml")
partlist2 = soup.find('div', class_ = 'part_list_2').find_all('h3')
for idx, news in enumerate(partlist2):
    print("%s, %s" %(idx, news.get_text()))
```

## 2. 取出現在時間兩小時內的新聞

```python
browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")

for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html.parser")
localtime = time.localtime()
if localtime.tm_mon < 10:
    zfillmon = str(localtime.tm_mon).zfill(2)
localtime = str(localtime.tm_year) + zfillmon + str(localtime.tm_mday) + str(localtime.tm_hour) + str(localtime.tm_min)

for idx, d in enumerate(soup.find(class_="part_list_2").find_all('h3')):

    time = d.find(class_ = "date").string
    formattime = time.replace("/", "").replace(":", "").replace(" ","")
    if int(formattime) > int(localtime) - 300:
        # print("%d , %d ------"%(int(formattime), int(localtime) - 300))
        print(d.get_text())

browser.quit()
```
