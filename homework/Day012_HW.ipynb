### ① 取出今天所有的發文

```python
'''
Your Code
'''
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "lxml")
dict = {}
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find(class_="tag").text, d.find_all('a')[-1].text)
    
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
'''
Your Code
'''
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "lxml")
dict = {}
for d in soup.find(class_="part_list_2").find_all('h3'):
    if d.find(class_="tag").text not in dict:
        dict.setdefault(d.find(class_="tag").text, 1)
#         dict[d.find(class_="tag").text] = 1
    elif d.find(class_="tag").text in dict:
        dict[d.find(class_="tag").text] += 1
print(dict)
```

### ③ 哪一個類別的文章最多

```python
'''
Your Code
'''
print(max(dict))
```
